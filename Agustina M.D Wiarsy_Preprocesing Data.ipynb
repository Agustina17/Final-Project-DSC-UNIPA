```python
#Handling Missing Value
data['Tenure'] = data['Tenure'].fillna(data['Tenure'].mean())
data['WarehouseToHome'] = data['WarehouseToHome'].fillna(data['WarehouseToHome'].mode()[0])
data['HourSpendOnApp'] = data['HourSpendOnApp'].fillna(data['HourSpendOnApp'].mode()[0])
data['OrderIncreaseFromLastYear'] = data['OrderIncreaseFromLastYear'].fillna(data['OrderIncreaseFromLastYear'].mode()[0])
data['CouponUsed'] = data['CouponUsed'].fillna(data['CouponUsed'].mode()[0])
data['OrderCount'] = data['OrderCount'].fillna(data['OrderCount'].mode()[0])
data['DaySinceLastOrder'] = data['DaySinceLastOrder'].fillna(data['DaySinceLastOrder'].mode()[0])
```

```python
data.isnull().sum()
```

```python
#Centering
data['Tenure_scaled'] = data['Tenure']-data['Tenure'].mean()

centering = data[['Tenure','Tenure_scaled']]
centering.head()
```

```python
#Standarisasi
data['CashbackAmount_scaled'] = (data['CashbackAmount']-data['CashbackAmount'].mean())/data['CashbackAmount'].std()

standarisasi = data[['CashbackAmount','CashbackAmount_scaled']]
standarisasi.head()
```
